```python
from transformers import pipeline
import textwrap

# Load the question-answering pipeline
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

def split_text(text, max_length=500):
    """ Splits long text into smaller chunks for better model processing. """
    return textwrap.wrap(text, max_length)

def get_answers(questions, context):
    """ Answers multiple questions from a given context, handling long text. """
    answers = {}
    
    # Split context if it's too long
    context_chunks = split_text(context)
    
    for question in questions:
        best_answer = ""
        best_score = 0
        
        # Process each chunk separately
        for chunk in context_chunks:
            result = qa_pipeline(question=question, context=chunk)
            if result["score"] > best_score:
                best_score = result["score"]
                best_answer = result["answer"]
        
        answers[question] = best_answer
    return answers

# Example Context (LONG PASSAGE)
context = """The Great Wall of China is one of the most famous landmarks in the world. 
It was built to protect Chinese states from invasions and raids. The wall stretches over 13,000 miles 
and was built during different dynasties, most notably during the Ming Dynasty. The construction 
involved millions of workers, including soldiers, peasants, and prisoners. The wall is also a symbol 
of China's historical strength and determination, attracting millions of tourists every year."""

# Example Questions
questions = [
    "Why was the Great Wall of China built?",
    "How long is the Great Wall?",
    "Which dynasty built most of it?",
    "Who helped in constructing the wall?",
    "What is the Great Wall a symbol of?"
]

# Get Answers
answers = get_answers(questions, context)
for q, a in answers.items():
    print(f"Q: {q}\nA: {a}\n")
```

```python
import pickle
from transformers import AutoModelForQuestionAnswering, AutoTokenizer

# Load model and tokenizer from Hugging Face
model = AutoModelForQuestionAnswering.from_pretrained("deepset/roberta-base-squad2")
tokenizer = AutoTokenizer.from_pretrained("deepset/roberta-base-squad2")

# Save model as a pickle file
with open("qa_model.pkl", "wb") as model_file:
    pickle.dump(model, model_file)

with open("qa_tokenizer.pkl", "wb") as tokenizer_file:
    pickle.dump(tokenizer, tokenizer_file)

print("Model and tokenizer saved as pickle files!")
```
